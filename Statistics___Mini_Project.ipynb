```python
import pandas as pd
from scipy import stats
```

```python
df.columns = df.columns.str.strip()
```

```python
print("========== STEP 5: HYPOTHESIS TESTING ==========\n")
```

```python
# INDEPENDENT T-TEST: Do males and females spend differently?

print("1. Business Question: Do males and females spend differently?")
print("   Test: Independent t-test")

male_spend = df[df['Gender'] == 'Male']['MonthlySpend']
female_spend = df[df['Gender'] == 'Female']['MonthlySpend']

t_stat, p_val_t = stats.ttest_ind(male_spend, female_spend)
print(f"   Result -> T-statistic: {t_stat:.4f}, P-value: {p_val_t:.4f}")
if p_val_t < 0.05:
    print("   Conclusion: Yes, there is a significant difference in spend between males and females.\n")
else:
    print("   Conclusion: No, there is no significant difference in spend between males and females.\n")

print("-" * 60)
```

```python
# ONE-WAY ANOVA: Does education level impact average monthly spend?
print("\n2. Business Question: Does education level impact average monthly spend?")
print("   Test: One-way ANOVA")

education_groups = [group['MonthlySpend'].values for name, group in df.groupby('Education')]
f_stat_edu, p_val_edu = stats.f_oneway(*education_groups)
print(f"   Result -> F-statistic: {f_stat_edu:.4f}, P-value: {p_val_edu:.4f}")
if p_val_edu < 0.05:
    print("   Conclusion: Yes, education level significantly impacts monthly spend.\n")
else:
    print("   Conclusion: No, education level does not significantly impact monthly spend.\n")

print("-" * 60)
```

```python
# CHI-SQUARE TEST: Is marital status related to the number of pets owned?
print("\n3. Business Question: Is marital status related to the number of pets owned?")
print("   Test: Chi-square test")

contingency_table = pd.crosstab(df['Married'], df['NumPets'])

chi2_stat, p_val_chi2, dof, expected = stats.chi2_contingency(contingency_table)
print(f"   Result -> Chi2-statistic: {chi2_stat:.4f}, P-value: {p_val_chi2:.4f}")
if p_val_chi2 < 0.05:
    print("   Conclusion: Yes, marital status is significantly related to the number of pets owned.\n")
else:
    print("   Conclusion: No, marital status is not significantly related to the number of pets owned.\n")

print("-" * 60)
```

```python
# CORRELATION: Are older people less active?

print("\n4. Business Question: Are older people less active?")
print("   Test: Correlation (Age vs DaysSinceLastInteraction)")

corr_stat, p_val_corr = stats.pearsonr(df['Age'], df['DaysSinceLastInteraction'])
print(f"   Result -> Correlation Coefficient: {corr_stat:.4f}, P-value: {p_val_corr:.4f}")

if p_val_corr < 0.05 and corr_stat > 0:
    print("   Conclusion: Yes, older people are significantly less active (as Age goes up, Days Since Last Interaction goes up).\n")
elif p_val_corr < 0.05 and corr_stat < 0:
    print("   Conclusion: No, older people are actually MORE active.\n")
else:
    print("   Conclusion: No significant relationship between age and activity level.\n")

print("-" * 60)
```

```python
# 5. ANOVA: Does state-wise spend vary significantly?
print("\n5. Business Question: Does state-wise spend vary significantly?")
print("   Test: ANOVA")

state_groups = [group['MonthlySpend'].values for name, group in df.groupby('State')]

f_stat_state, p_val_state = stats.f_oneway(*state_groups)
print(f"   Result -> F-statistic: {f_stat_state:.4f}, P-value: {p_val_state:.4f}")
if p_val_state < 0.05:
    print("   Conclusion: Yes, state-wise spend varies significantly.\n")
else:
    print("   Conclusion: No, state-wise spend does not vary significantly.\n")

print("\n" + "="*60)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.stats.api as sms

df.columns = df.columns.str.strip()

print("========== STEP 6: FORMAL HYPOTHESIS TESTING ==========\n")
print("BUSINESS QUESTION: Do males and females spend differently?\n")
print("-" * 65)

print("1. DEFINE HYPOTHESES:")
print("   Null (H0): There is NO difference in average monthly spend between males and females.")
print("   Alternate (H1): There IS a significant difference in average monthly spend between males and females.")

print("\n2. TEST SELECTION:")
print("   Test Chosen: Independent Samples t-test.")
print("   Reason: We are comparing the means of a continuous numeric variable (Spend) across exactly two categorical groups (Male/Female).")

print("\n3. ASSUMPTIONS CHECK:")
male_spend = df[df['Gender'] == 'Male']['MonthlySpend'].dropna()
female_spend = df[df['Gender'] == 'Female']['MonthlySpend'].dropna()

print("   - Independence: Assumed met (each customer ID represents a distinct, independent person).")

stat_m, p_m = stats.shapiro(male_spend)
stat_f, p_f = stats.shapiro(female_spend)
print(f"   - Normality (Shapiro-Wilk): Male p-value = {p_m:.4f}, Female p-value = {p_f:.4f}")
if p_m > 0.05 and p_f > 0.05:
    print("     -> Data appears normally distributed (Assumption met).")
else:
    print("     -> Note: P-value < 0.05 indicates deviation from perfect normality. However, with large sample sizes (N>30), the t-test remains robust due to the Central Limit Theorem.")

stat_lev, p_lev = stats.levene(male_spend, female_spend)
print(f"   - Homogeneity of Variance (Levene's Test): p-value = {p_lev:.4f}")
equal_var_assumed = p_lev > 0.05
if equal_var_assumed:
    print("     -> Variances are equal (Assumption met). We will use the standard Student's t-test.")
else:
    print("     -> Variances are NOT equal. We will use Welch's t-test to correct for this.")

print("\n4. INTERPRET P-VALUES & CONFIDENCE INTERVALS:")

t_stat, p_val = stats.ttest_ind(male_spend, female_spend, equal_var=equal_var_assumed)

cm = sms.CompareMeans(sms.DescrStatsW(male_spend), sms.DescrStatsW(female_spend))
lower_ci, upper_ci = cm.tconfint_diff(usevar='pooled' if equal_var_assumed else 'unequal')

print(f"   T-Statistic: {t_stat:.4f}")
print(f"   P-Value: {p_val:.4f}")
print(f"   95% Confidence Interval for Difference in Means: ({lower_ci:.2f}, {upper_ci:.2f})")

print("\n   FINAL CONCLUSION:")
if p_val < 0.05:
    print("   -> We REJECT the null hypothesis.")
    print(f"   -> Interpretation: There is a statistically significant difference. We are 95% confident that the true difference in average monthly spend between males and females falls between ${lower_ci:.2f} and ${upper_ci:.2f}.")
else:
    print("   -> We FAIL TO REJECT the null hypothesis.")
    print(f"   -> Interpretation: There is no statistically significant difference in spend. Because the confidence interval ({lower_ci:.2f} to {upper_ci:.2f}) includes the number ZERO, the difference could very well be nothing.")
```
